```python
#calculator App
num1 = int(input("Enter first Number"))
num2 = int(input("Enter Second Number"))
operator = input("Enter your Operator")

if operator == "+":
    print(num1+num2)
elif operator == "-":
    print(num1-num2)
elif operator == "/":
    print(num1/num2)
elif operator == "*":
    print(num1*num2)
else:
    print("invalid operator")
           
```
