```python
#Marksheet 

Eng = int(input("Enter your EngLish marks"))
Math = int (input("Enter your Math marks"))
Comp = int(input("Enter your Computer Marks"))
Sci = int(input("Enter your Science marks"))
Sindhi = int(input("Enter your Sindhi marks"))
             
ObtainMarks = Eng+Math+Comp+Sci+Sindhi

print("Your Obain marks is : ", ObtainMarks)
perc = ObtainMarks / 500 * 100
print ("Your Percentage is : ", perc )
      
if perc >= 90 and perc <= 100:
    print ("A+ Grade")
elif perc >= 80 and perc < 80:
    print("A Grade")
elif perc >= 70 and perc < 80:
    print ("B Grade")
elif per >=60 and perc < 70:
    print("C Grade")
else:
    print(" E Grade,You are Fail")



```
